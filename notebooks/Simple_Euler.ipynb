```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12, 8)


def Euler_step(t,y,func,step=.1):
    ynew=y+step*func(t,y)
    return ynew

def Euler(func,step=.1,final_time=1,t0=0,y0=1):
    steps=np.int(final_time/step)
    sol=np.zeros(steps+1)
    #times=np.linspace(t0,final_time,steps)
    sol[0]=y0
    time=t0
    yold=y0
    for i in range(1,steps+1):
        ynew=Euler_step(time,yold,func,step)
        sol[i]=ynew
        yold=ynew
        time+=step
    return sol

def interest(t,y):
    return y

for step in (.5,.25,.125,.0625):
    ans=Euler(interest,step=step)
    time=np.linspace(0,1,np.size(ans))
    plt.plot(time,ans,marker='o')

plt.plot(time,np.exp(time))
plt.show()



```
