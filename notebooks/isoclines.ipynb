```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12, 8)

def tplus2y(t,y):
    return t+2*y

def isocline(func,t=np.array([-4.0,4.0]),y=np.array([-4.0,4.0])):
    tgrid=np.linspace(t[0],t[1],20)
    ygrid=np.linspace(y[0],y[1],20)
    TP,YP = np.meshgrid(tgrid,ygrid)
    temp=np.arctan(func(TP,YP))
    plt.quiver(TP,YP,np.cos(temp),np.sin(temp))
    #plt.show()
fig,ax=plt.subplots()
ax.set_xlim(-4,4)
y=np.linspace(-4,4,40)
isocline(tplus2y)
for c in range(-5,5):
    t=c-2*y
    plt.plot(t,y,label='$c=$'+str(c))
    
#ax.set_xlim(-4,4)
plt.legend()
plt.show()




```
