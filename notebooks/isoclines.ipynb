```python
import numpy as np
import matplotlib.pyplot as plt

def tplus2y(t,y):
    return t+2*y

def isocline(func,t=np.array([-4.0,4.0]),y=np.array([-4.0,4.0])):
    tgrid=np.linspace(t[0],t[1],20)
    ygrid=np.linspace(y[0],y[1],20)
    TP,YP = np.meshgrid(tgrid,ygrid)
    temp=np.arctan(func(TP,YP))
    plt.quiver(TP,YP,np.cos(temp),np.sin(temp))
    plt.show()

isocline(tplus2y)



```
